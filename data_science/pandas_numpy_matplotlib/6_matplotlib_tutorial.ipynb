```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# Practice with tutorial: https://www.youtube.com/watch?v=3Xc3CA655Y4
```

```python
x = np.arange(0,10,2)
y = [1, 3, 2, 5, 6.5]
z = np.sin(x)
x_1 = np.linspace(0, 8, 20)

plt.figure(figsize=(10, 5)) # dpi=300) # dpi=pixels per inch
plt.title('XnameX', fontdict={'fontsize': 20, 'fontname' : 'Comic Sans MS'})
plt.xlabel('x label', fontdict={'fontsize': 15, 'fontname' : 'Comic Sans MS'})
plt.ylabel('y label')
plt.yticks(np.arange(-1, 1.2, 0.2))
plt.ylim((-1.1, 1.1))

plt.plot(x[:4], z[:4], label='some function', color='r', linewidth=3)
plt.plot(x[3:], z[3:], label='some function but blue', color='b', linewidth=5)
plt.plot(x_1, np.sin(x_1), 'r:', label='sin function', linewidth=3)
plt.axhline(linewidth=1, color='black', linestyle='-.')
plt.legend()
plt.savefig('matplotlib_graph_1', dpi=300)
plt.show()
```

```python
labels = ['A', 'B', 'C']
values = [1, 2, 3]

bar_plot = plt.bar(labels, values)
bar_plot[0].set_hatch('/')
bar_plot[1].set_hatch('O')
bar_plot[2].set_hatch('*')

plt.show()
```

```python
gas = pd.read_csv('datasets\gas_prices.csv')
gas
```

```python
plt.figure(figsize=(15,8))

for country in gas:
    plt.plot(gas.Year, gas[country], label=str(country), marker='.')

plt.xticks(np.arange(1990, 2010, 3))
plt.ylim(0,8)
plt.ylabel('Price in USD', fontdict={'fontsize':13})
plt.xlabel('Year', fontdict={'fontsize':13})
plt.title('Change of gas prices', fontdict={'fontsize':15})
plt.grid()

plt.legend()

plt.show()
```

```python
fifa = pd.read_csv('datasets/fifa_data.csv')
fifa.head()
```

```python
bins = np.arange(40, 110, 10)
plt.hist(fifa.Overall, bins=bins)
plt.xticks(bins)

plt.title('fifa player scores')
plt.xlabel('score')
plt.ylabel('n of players')
plt.show()
```

```python
fifa = fifa.rename(columns={'Preferred Foot' : 'pref_foot'})
lefties = fifa.query('pref_foot == "Left"').count()[0]
righties = fifa.count()[0] - lefties

plt.title('XXX')
labels= ['Left', 'Right']
plt.pie([lefties, righties], labels=labels, autopct='%.2f %%', textprops=dict(color="w"))

plt.show()
```

```python
fifa.Weight.describe()
print(fifa.Weight.str.contains("lbs").count() == fifa.Weight.count())
```

```python
print(fifa.Weight.isna().sum())
fifa = fifa[fifa['Weight'].notna()]
print(fifa.Weight.isna().sum())

fifa.Weight = fifa.Weight.str[:-3].astype('int')
fifa.Weight
```

```python
fifa['Weight'] = fifa['Weight']*0.45359237
fifa.Weight.describe()
```

```python
w_light = fifa.query('Weight < 70')
w_medium = fifa.query('70 <= Weight < 80')
w_heavy = fifa.query('80 <= Weight')
print(len(w_light), len(w_medium), len(w_heavy))
```

```python
explode = [0.03, 0.03, 0.03]
plt.pie([len(w_light), len(w_medium), len(w_heavy)], labels=['light', 'medium', 'heavy'], autopct='%.2f %%', textprops=dict(color="w"), explode=explode)

plt.title('Distribution of players weight', color='w')
plt.show()
```
