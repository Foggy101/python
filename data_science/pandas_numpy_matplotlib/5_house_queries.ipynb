```python
df = df.rename(columns={'Posted On' : 'posted_date', 'BHK' : 'n_rooms', 'Rent' : 'rent', 'Size' : 'size_ft', 'Floor' : 'floor', 'Area Type' : 'area_type',
                        'Area Locality' : 'area_locality', 'City' : 'city', 'Furnishing Status' : 'furnish', 'Tenant Preferred' : 'pref_tenant',
                        'Bathroom' : 'n_bathroom', 'Point of Contact' : 'seller_type'})
```

```python
df.dtypes
```

```python
df.posted_date = pd.to_datetime(df.posted_date)
```

```python
today_date = datetime.today().strftime('%Y-%m-%d')
df.insert(1, 'time_since_posted', pd.to_datetime(today_date) - df.posted_date)
df = df.drop(['posted_date'], axis='columns')
df
```

```python
df.insert(4, 'size_m2', (df.size_ft * 0.09290304))
df.size_m2 = df.size_m2.apply(np.floor).astype('int')
df = df.drop(['size_ft'], axis='columns')
df
```

```python
df.pref_tenant = df.pref_tenant.str.replace('/', ',').str.lower()
df.seller_type = df.seller_type.apply(lambda x: x.split()[-1])
df.insert(4, 'price_per_m2', (df.rent / df.size_m2).apply(np.floor))
df
```

```python
df.insert(6, 'max_floor', df.floor.apply(lambda x: x.split()[-1]))
df
```

```python
df.floor = df.floor.apply(lambda x: x.split()[0])
df = df.replace({'floor' : {'Upper' : -1, 'Lower' : -2, 'Ground' : 0}})
df.floor = df.floor.astype('int')
df.floor.unique()
```

# QUERY TASKS
## seller owner, under 150/m2, last 30 days
## upper 50% of max floor
## underground, bachelor +- family, city starting with A-H

```python
df.insert(1, 'time_query', df.time_since_posted.astype('str').apply(lambda x: x.split()[0]).astype('int'))
```

```python
df.query('seller_type == "Owner" & price_per_m2 < 150 & time_query < 90')
```

```python
df.query('max_floor == "Ground"')
```

```python
df.max_floor = df.max_floor.replace('Ground', 0)
df.max_floor = df.max_floor.astype('int')
df.max_floor.unique()
```

```python
df = df.rename(columns={'floor' : 'which_floor'})
df.query('which_floor/max_floor >= 1 & max_floor > 15')
```

```python
df.query('which_floor < 0 & (pref_tenant == "bachelors" | pref_tenant == "bachelors,family")')
```
